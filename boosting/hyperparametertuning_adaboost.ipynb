```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
```

### creating a dataset using the make circles

```python
x, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter')
```

### using adaboost classifier from scikitlearn library

```python
adb = AdaBoostClassifier()
```

### calculating the cross validation score with scoring as accuracy

```python
score = np.mean(cross_val_score(adb, x, y, cv=10, scoring="accuracy"))
```

```python
print(score)
```

```python
adb.fit(x,y)
```

```python
y_pred = adb.predict(x)
```

```python
accuracy_score(y,y_pred)
```

### using grid search cv to get the best parameters for hypertuning

```python
grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
grid['algorithm'] = ['SAMME', 'SAMME.R']
```

```python
reg = GridSearchCV(AdaBoostClassifier(), param_grid=grid, cv = 10, n_jobs=-1, verbose=1)
```

```python
reg.fit(x,y)
```

### best score using Grid search cv
